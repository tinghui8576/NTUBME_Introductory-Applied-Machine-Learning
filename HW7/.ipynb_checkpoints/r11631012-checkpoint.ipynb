### Problem 2 (Support vector machine)

```python
import scipy.io as sio
mnist_data = sio.loadmat("07HW2_digit.mat", mdict=None, appendmat=True)
```

```python
training_set = []
testing_set = []
for samples in mnist_data:
    if 'train' in samples:
        batch = mnist_data[samples].astype(float)
        training_set.append(batch)
    if 'test' in samples:
        batch = mnist_data[samples].astype(float)
        testing_set.append(batch)
training_set = np.array(training_set)
testing_set = np.array(testing_set)
print(training_set.shape)
plt.imshow(testing_set[0, 0].reshape(28, 28), cmap='gray')
```

```python
def normalize(arr):
    arr -= arr.min()
    return arr / arr.max()


training_set = normalize(training_set)
testing_set = normalize(testing_set)
```

(a)

```python
from sklearn import svm

# hyperparamter
gamma = [pow(2, i) for i in range(-14, -5)]
c = [pow(2, i) for i in range(-5, 4)]
train_class = np.repeat(range(2), training_set[0].shape[0])
train_data = np.append(training_set[0], training_set[1], axis=0)
test_class = np.repeat(range(2), testing_set[0].shape[0])
test_data = np.append(testing_set[0], testing_set[1], axis=0)

# I used test data rather than cross-validation to test the SVM model ACC
acc_table = np.zeros((len(gamma), len(c)))
for i, hyp_g in enumerate(gamma):
    for j, hyp_c in enumerate(c):
        predictor = svm.SVC(gamma=hyp_g, C=hyp_c, kernel='rbf')
        # predictor = svm.OneClassSVM(gamma=hyp_g, C=hyp_c, kernel='rbf')
        predictor.fit(train_data, train_class)
        result = predictor.predict(test_data)
        accuracy = np.sum(np.equal(result, test_class)) / len(test_class)
        acc_table[i, j] = accuracy
        # print(f'With gamma={hyp_g} and c={hyp_c} accuracy is {accuracy}')
print(acc_table)
```

On seeing the result, the model can reach 100% acc with many different hyp pair.
At the same time, we should chose the best acc with maximize c and minimize gamma.
Therefor the best hyp pair is **(gamma, c) = (pow(2, -14), pow(2, 3))**

(b)

```python
# linear
acc_table = np.zeros((len(gamma), len(c)))
for i, hyp_g in enumerate(gamma):
    for j, hyp_c in enumerate(c):
        predictor = svm.SVC(gamma=hyp_g, C=hyp_c, kernel='linear')
        predictor.fit(train_data, train_class)
        result = predictor.predict(test_data)
        accuracy = np.sum(np.equal(result, test_class)) / len(test_class)
        acc_table[i, j] = accuracy
print(acc_table)
```

```python
# polynomial
for degree in range(1, 4):
    acc_table = np.zeros((len(gamma), len(c)))
    for i, hyp_g in enumerate(gamma):
        for j, hyp_c in enumerate(c):
            predictor = svm.SVC(gamma=hyp_g, C=hyp_c,
                                degree=degree, kernel='poly')
            predictor.fit(train_data, train_class)
            result = predictor.predict(test_data)
            accuracy = np.sum(np.equal(result, test_class)) / len(test_class)
            acc_table[i, j] = accuracy
    print(f'degree={degree}')
    print(acc_table)
```

For linear kernel, every pair reach an acc of 100%. <br>
Using polynomial kernel with different degree (1-3, i had tried others, performance not well). <br>
To limiting the compute complexity, the best hyps is **(degree, gamma, c) = (1, pow(2, -14), pow(2, 3))**.

(C)

```python
# Implement multiple-class SVM with one to the rest method
def my_svm_train(training_set, hyp_g, hyp_c, kernel='rbf', degree=3):
    # training_set should be a 3-D matrix with (num_of_cls, num_of_spl, dim_of_spl)
    svm_model = []
    for i in range(training_set.shape[0]):
        # print(f'On training class {i} svm classifier.')
        train_target = training_set[i]
        other_set = np.delete(training_set, i, axis=0)
        train_other = other_set.reshape(
            other_set.shape[0]*other_set.shape[1], other_set.shape[2])
        train_data = np.append(train_target, train_other, axis=0)

        # if predicted result is -1, is not the index svm model's class.
        class_target = np.repeat(i, training_set[i].shape[0])
        class_other = np.repeat(-1, train_other.shape[0])
        class_train = np.append(class_target, class_other)
        if kernel == 'poly':
            predictor = svm.SVC(gamma=hyp_g, C=hyp_c,
                                kernel=kernel, degree=degree)
        else:
            predictor = svm.SVC(gamma=hyp_g, C=hyp_c, kernel=kernel)
        predictor.fit(train_data, class_train)
        svm_model.append(predictor)
    # Return svm_model[index] can predict class of index .
    return svm_model
```

```python
# my multiple-class SVM prediction
def my_svm_predict(svm_model, test_data):
    result = np.zeros(test_data.shape[0])
    for index, predictor in enumerate(svm_model):
        class_result = predictor.predict(test_data)
        result[class_result == np.repeat(index, len(class_result))] = index
        # print()
    return result
```

```python
gamma = [pow(2, i) for i in range(-7, -5)]
c = [pow(2, i) for i in range(-3, 4)]
test_data = testing_set.reshape(
    testing_set.shape[0]*testing_set.shape[1], testing_set.shape[2])
test_label = np.repeat(range(10), testing_set.shape[1])
```

```python
best_hyp = [0.0, 0.0, 0.0]
for i, hyp_g in enumerate(gamma):
    for j, hyp_c in enumerate(c):
        number_svm = my_svm_train(training_set, hyp_g, hyp_c, kernel='rbf')
        my_result = my_svm_predict(number_svm, test_data)
        accuracy = np.sum(np.equal(my_result, test_label)) / len(test_label)
        if accuracy > best_hyp[2]:
            best_hyp[0] =hyp_g
            best_hyp[1] =hyp_c
            best_hyp[2] = accuracy
print(f'Best accuracy of rbf is {best_hyp[2]}, ')
print(f'  with gamma = {best_hyp[0]} and c = {best_hyp[1]}')
```

```python
best_hyp = [0.0, 0.0, 0.0]
for i, hyp_g in enumerate(gamma):
    for j, hyp_c in enumerate(c):
        number_svm = my_svm_train(training_set, hyp_g, hyp_c, kernel='linear')
        my_result = my_svm_predict(number_svm, test_data)
        accuracy = np.sum(np.equal(my_result, test_label)) / len(test_label)
        if accuracy > best_hyp[2]:
            best_hyp[0] =hyp_g
            best_hyp[1] =hyp_c
            best_hyp[2] = accuracy
print(f'Best accuracy of linear is {best_hyp[2]}, ')
print(f'  with gamma = {best_hyp[0]} and c = {best_hyp[1]}')
```

```python
for degree in range(2, 5):
    best_hyp = [0.0, 0.0, 0.0]
    for i, hyp_g in enumerate(gamma):
        for j, hyp_c in enumerate(c):
            number_svm = my_svm_train(
                training_set, hyp_g, hyp_c, kernel='poly', degree=degree)
            my_result = my_svm_predict(number_svm, test_data)
            accuracy = np.sum(np.equal(my_result, test_label)
                              ) / len(test_label)
            if accuracy > best_hyp[2]:
                best_hyp[0] =hyp_g
                best_hyp[1] =hyp_c
                best_hyp[2] = accuracy
    print(f'Best accuracy of ploy-{degree} is {best_hyp[2]}, ')
    print(f'  with gamma = {best_hyp[0]} and c = {best_hyp[1]}')
```
