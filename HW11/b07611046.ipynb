# Problem 1 (k-means)

```python
import scipy.io as sio
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = sio.loadmat("11HW1_KmeanData.mat", mdict=None, appendmat=True)
X = np.float64(data['X']).T
Y = np.float64(data['Y']).T

plt.figure(figsize=(12,8))
#k=4
#x
plt.subplot(231)
kmeans = KMeans(n_clusters=4).fit(X)
predict_x = kmeans.predict(X)
plt.title('Kmeans for X, K = 4')
plt.scatter(X.T[0], X.T[1], c=predict_x, alpha=0.5)
plt.scatter(kmeans.cluster_centers_.T[0], kmeans.cluster_centers_.T[1], marker = 'x', s=40, c='black')
plt.ylim([-5, 6])
#y
plt.subplot(234)
kmeans = KMeans(n_clusters=4).fit(Y)
predict_y = kmeans.predict(Y)
plt.title('Kmeans for Y, K = 4')
plt.scatter(Y.T[0], Y.T[1], c=predict_y, alpha=0.5)
plt.scatter(kmeans.cluster_centers_.T[0], kmeans.cluster_centers_.T[1], marker = 'x', s=40, c='black')
plt.ylim([-5, 5])

#k=5
#x
plt.subplot(232)
kmeans = KMeans(n_clusters=5).fit(X)
predict_x = kmeans.predict(X)
plt.title('Kmeans for X, K = 5')
plt.scatter(X.T[0], X.T[1], c=predict_x, alpha=0.5)
plt.scatter(kmeans.cluster_centers_.T[0], kmeans.cluster_centers_.T[1], marker = 'x', s=40, c='black')
plt.ylim([-5, 6])
#y
plt.subplot(235)
kmeans = KMeans(n_clusters=5).fit(Y)
predict_y = kmeans.predict(Y)
plt.title('Kmeans for Y, K = 5')
plt.scatter(Y.T[0], Y.T[1], c=predict_y, alpha=0.5)
plt.scatter(kmeans.cluster_centers_.T[0], kmeans.cluster_centers_.T[1], marker = 'x', s=40, c='black')
plt.ylim([-5, 5])

#k=6
#x
plt.subplot(233)
kmeans = KMeans(n_clusters=6).fit(X)
predict_x = kmeans.predict(X)
plt.title('Kmeans for X, K = 6')
plt.scatter(X.T[0], X.T[1], c=predict_x, alpha=0.5)
plt.scatter(kmeans.cluster_centers_.T[0], kmeans.cluster_centers_.T[1], marker = 'x', s=40, c='black')
plt.ylim([-5, 6])
#y
plt.subplot(236)
kmeans = KMeans(n_clusters=6).fit(Y)
predict_y = kmeans.predict(Y)
plt.title('Kmeans for Y, K = 6')
plt.scatter(Y.T[0], Y.T[1], c=predict_y, alpha=0.5)
plt.scatter(kmeans.cluster_centers_.T[0], kmeans.cluster_centers_.T[1], marker = 'x', s=40, c='black')
plt.ylim([-5, 5])

plt.show()
```

# Problem 2 (k-means)

## (a) 

```python
grey = cv2.imdecode(np.fromfile('11HW2_PatchPanelsGrey.jpeg',dtype=np.uint8),1)
color = cv2.imdecode(np.fromfile('11HW2_PatchPanels.jpg',dtype=np.uint8),1)

Image_grey = np.reshape(grey/255, [np.size(grey, 0)*np.size(grey, 1), np.size(grey, 2)])
Image_color = np.reshape(color, [np.size(color, 0)*np.size(color, 1), np.size(color, 2)])
```

```python
plt.figure(figsize=(8,20))
#k=2
#grey
plt.subplot(521)
kmeans = KMeans(n_clusters=2).fit(Image_grey)
pred_d = kmeans.cluster_centers_[kmeans.labels_]
c_d= kmeans.cluster_centers_
pred_d = np.reshape(pred_d, [np.size(grey, 0), np.size(grey, 1),np.size(grey, 2)])
plt.title('K = 2, Grey')
plt.axis('off')
plt.imshow(pred_d, cmap='gray')

#color
plt.subplot(522)
kmeans = KMeans(n_clusters=2).fit(Image_color)
c_pred_d = kmeans.cluster_centers_[kmeans.labels_]
c_c_d= kmeans.cluster_centers_
c_pred_d = np.reshape(c_pred_d, [np.size(color, 0), np.size(color, 1), np.size(color, 2)])
plt.title('K = 2, Color')
plt.axis('off')
plt.imshow(cv2.cvtColor(c_pred_d.astype(np.uint8), cv2.COLOR_BGR2RGB))


#k=5
#grey
plt.subplot(523)
kmeans = KMeans(n_clusters=5).fit(Image_grey)
pred_f = kmeans.cluster_centers_[kmeans.labels_]
c_f= kmeans.cluster_centers_
pred_f = np.reshape(pred_f, [np.size(grey, 0), np.size(grey, 1),np.size(grey, 2)])
plt.title('K = 5, Grey')
plt.axis('off')
plt.imsave( '5.jpg',pred_f)
plt.imshow(pred_f, cmap='gray')
#color
plt.subplot(524)
kmeans = KMeans(n_clusters=5).fit(Image_color)
c_pred_f = kmeans.cluster_centers_[kmeans.labels_]
c_c_f= kmeans.cluster_centers_
c_pred_f = np.reshape(c_pred_f, [np.size(color, 0), np.size(color, 1), np.size(color, 2)])
plt.title('K = 5, Color')
plt.axis('off')
plt.imshow(cv2.cvtColor(c_pred_f.astype(np.uint8), cv2.COLOR_BGR2RGB))

#k=10
plt.subplot(525)
kmeans = KMeans(n_clusters=10).fit(Image_grey)
pred_o = kmeans.cluster_centers_[kmeans.labels_]
c_o= kmeans.cluster_centers_
pred_o = np.reshape(pred_o, [np.size(grey, 0), np.size(grey, 1),np.size(grey, 2)])
plt.title('K = 10, Grey')
plt.axis('off')
plt.imsave('10.jpg', pred_o)
plt.imshow(pred_o, cmap='gray')
#color
plt.subplot(526)
kmeans = KMeans(n_clusters=10).fit(Image_color)
c_pred_o = kmeans.cluster_centers_[kmeans.labels_]
c_c_o= kmeans.cluster_centers_
c_pred_o = np.reshape(c_pred_o, [np.size(color, 0), np.size(color, 1), np.size(color, 2)])
plt.title('K = 10, Color')
plt.axis('off')
plt.imshow(cv2.cvtColor(c_pred_o.astype(np.uint8), cv2.COLOR_BGR2RGB))

#k=15
#grey
plt.subplot(527)
kmeans = KMeans(n_clusters=15).fit(Image_grey)
pred_q = kmeans.cluster_centers_[kmeans.labels_]
c_q= kmeans.cluster_centers_
pred_q = np.reshape(pred_q, [np.size(grey, 0), np.size(grey, 1),np.size(grey, 2)])
plt.title('K = 15, Grey')
plt.axis('off')
plt.imshow(pred_q, cmap='gray')
#color
plt.subplot(528)
kmeans = KMeans(n_clusters=15).fit(Image_color)
c_pred_q = kmeans.cluster_centers_[kmeans.labels_]
c_c_q= kmeans.cluster_centers_
c_pred_q = np.reshape(c_pred_q, [np.size(color, 0), np.size(color, 1), np.size(color, 2)])
plt.title('K = 15, Color')
plt.axis('off')
plt.imshow(cv2.cvtColor(c_pred_q.astype(np.uint8), cv2.COLOR_BGR2RGB))

#k=20
plt.subplot(529)
kmeans = KMeans(n_clusters=20).fit(Image_grey)
pred_v = kmeans.cluster_centers_[kmeans.labels_]
c_v= kmeans.cluster_centers_
pred_v = np.reshape(pred_v, [np.size(grey, 0), np.size(grey, 1),np.size(grey, 2)])
plt.title('K = 20, Grey')
plt.axis('off')
plt.imshow(pred_v, cmap='gray')
#color
plt.subplot(5,2,10)
kmeans = KMeans(n_clusters=20).fit(Image_color)
c_pred_v = kmeans.cluster_centers_[kmeans.labels_]
c_c_v= kmeans.cluster_centers_
c_pred_v = np.reshape(c_pred_v, [np.size(color, 0), np.size(color, 1), np.size(color, 2)])
plt.title('K = 20, Color')
plt.axis('off')
plt.imshow(cv2.cvtColor(c_pred_v.astype(np.uint8), cv2.COLOR_BGR2RGB))

plt.show()
```

## (b)

```python
print('grey color image:')
#compression size: p*m*n+8*2^p*3 
#k=2, 2**1=2
s = np.size(grey, 0) * np.size(grey, 1) * np.size(grey, 2)*8
s_d =  np.size(c_d, 1) * np.size(c_d, 0) * 8 + np.size(grey, 0) * np.size(grey, 1) * 1
print(s,s_d)
com_d = s/s_d
print("Compression ration for k = 2: ",com_d)
# k=5, 2**3=8
s = np.size(grey, 0) * np.size(grey, 1) * np.size(grey, 2)*8
s_f = np.size(c_f, 1) * np.size(c_f, 0) * 8 + np.size(grey, 0) * np.size(grey, 1) * 3
com_f = s/s_f
print("Compression ration for k = 5: ",com_f )
# k=10, 2**4=16
s = np.size(grey, 0) * np.size(grey, 1) * np.size(grey, 2)*8
s_o = np.size(c_o, 1) * np.size(c_o, 0) * 8 + np.size(grey, 0) * np.size(grey, 1) * 4
com_o = s/s_o
print("Compression ration for k = 10: ",com_o)
# k=15, 2**4=16
s = np.size(grey, 0) * np.size(grey, 1) * np.size(grey, 2)*8
s_q = np.size(c_q, 1) * np.size(c_q, 0) * 8 + np.size(grey, 0) * np.size(grey, 1) * 4
com_q = s/s_q
print("Compression ration for k = 15: ",com_q)
# k=20, 2**5=32
s = np.size(grey, 0) * np.size(grey, 1) * np.size(grey, 2)*8
s_v = np.size(c_v, 1) * np.size(c_v, 0) * 8 + np.size(grey, 0) * np.size(grey, 1) * 5
com_v = s/s_v
print("Compression ration for k = 20: ",com_v)
```

```python
print('true color image:')
#k=2, 2**1=2
s_c = np.size(color, 0) * np.size(color, 1) * np.size(color, 2)*8
s_d_c =  np.size(c_c_d, 1) * np.size(c_c_d, 0) * 8 + np.size(color, 0) * np.size(color, 1) * 1
com_d_c = s_c/s_d_c
print("Compression ration for k = 2: ", com_d_c)
#k=5, 2**3=8
s_c = np.size(color, 0) * np.size(color, 1) * np.size(color, 2)*8
s_f_c =  np.size(c_c_f, 1) * np.size(c_c_f, 0) * 8 + np.size(color, 0) * np.size(color, 1) * 3
com_f_c = s_c/s_f_c
print("Compression ration for k = 5: ",com_f_c)
#k=10, 2**4=16
s_c = np.size(color, 0) * np.size(color, 1) * np.size(color, 2)*8
s_o_c =  np.size(c_c_o, 1) * np.size(c_c_o, 0) * 8 + np.size(color, 0) * np.size(color, 1) * 4
com_o_c = s_c/s_o_c
print("Compression ration for k = 10: ",com_o_c)
#k=15, 2**4=16
s_c = np.size(color, 0) * np.size(color, 1) * np.size(color, 2)*8
s_q_c =  np.size(c_c_q, 1) * np.size(c_c_q, 0) * 8 + np.size(color, 0) * np.size(color, 1) *4
com_q_c = s_c/s_q_c
print("Compression ration for k = 15: ",com_q_c)
#k=20, 2**5=32
s_c = np.size(color, 0) * np.size(color, 1) * np.size(color, 2)*8
s_v_c =  np.size(c_c_v, 1) * np.size(c_c_v, 0) * 8 + np.size(color, 0) * np.size(color, 1) * 5
com_v_c = s_c/s_v_c
print("Compression ration for k = 20: ",com_v_c)
```
