```python
import scipy.io as io
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

# Problem 1

```python
# load data
data = io.loadmat('11HW1_KmeanData.mat')

# adjust the shape of numpy
data_Y = data['Y'].reshape((200, 1)) 
data_X = data['X'].reshape((200, 1))
data_point = np.concatenate((data_X, data_Y), axis=1) #horizontal concatenate

# train 
kmeans4 = KMeans(n_clusters=4)
kmeans5 = KMeans(n_clusters=5)
kmeans6 = KMeans(n_clusters=6)
kmeans4.fit(data_point);kmeans5.fit(data_point);kmeans6.fit(data_point)

#predict
y_k4 = kmeans4.predict(data_point)
y_k5 = kmeans5.predict(data_point)
y_k6 = kmeans6.predict(data_point)

# original data distribution
plt.scatter(data_point[:, 0], data_point[:, 1]) 
```

```python
plt.figure(figsize=(12,9))

plt.subplot(221)
plt.scatter(data_point[:, 0], data_point[:, 1], c=y_k4, cmap='viridis')
centers = kmeans4.cluster_centers_ #plot the center point
plt.scatter(centers[:, 0], centers[:, 1], c='black', alpha=0.5)
plt.title('The result of K=4')
plt.xticks([])
plt.yticks([])

plt.subplot(222)
plt.scatter(data_point[:, 0], data_point[:, 1], c=y_k5, cmap='viridis')
centers = kmeans5.cluster_centers_ #plot the center point
plt.scatter(centers[:, 0], centers[:, 1], c='black', alpha=0.5)
plt.title('The result of K=5')
plt.xticks([])
plt.yticks([])

plt.subplot(223)
plt.scatter(data_point[:, 0], data_point[:, 1], c=y_k6, cmap='viridis')
centers = kmeans6.cluster_centers_ #plot the center point
plt.scatter(centers[:, 0], centers[:, 1], c='black', alpha=0.5)
plt.title('The result of K=6')
plt.xticks([])
plt.yticks([])
plt.show()
```

# Problem 2

```python
k = ['2','5','10','15', '20'] 
rgb_img = plt.imread('11HW2_PatchPanels.jpg')       #(480, 640, 3), :92.4KB
gray_img = plt.imread('11HW2_PatchPanelsGrey.jpeg') #(665, 666, 3), :213KB
```

```python
X2 = (rgb_img/255).reshape(rgb_img.shape[0]*rgb_img.shape[1],3) # float64
X1 = (gray_img/255).reshape(gray_img.shape[0]*gray_img.shape[1],3)
num_iters = 10   # the number of repeating the process
```

```python
def runKMeans(X, initial_centroids, K, num_iters):
    idx = findClosestCentroids(X, initial_centroids)
    
    for i in range(num_iters):
        centroids = computeCentroids(X, idx, K)
        idx = findClosestCentroids(X, centroids)
    return centroids, idx

def computeCentroids(X, idx, K):
    m, n = X.shape
    centroids = np.zeros((K, n))
    count = np.zeros((K, 1))
    for i in range(m):
        index = int((idx[i]-1)[0])
        centroids[index, :] += X[i, :]
        count[index] += 1
    return centroids/count

def findClosestCentroids(X, centroids):
    K = centroids.shape[0]
    m = X.shape[0]
    
    idx = np.zeros((m, 1))
    temp = np.zeros((K, 1))
    
    for i in range(m):
        for j in range(K):
            temp[j] = np.sqrt(np.sum( (X[i, :] - centroids[j, :])**2 ))
        idx[i] = np.argmin(temp)+1
    return idx

def kMeansInit(X, K):
    m, n = X.shape
    centroids = np.zeros((K, n))
    
    for i in range(K):
        centroids[i] = X[np.random.randint(0, m+1), :]
    return centroids
```

```python
# for rgb_img
for i in k:
    initial_centroids = kMeansInit(X2, int(i))
    centroids2, idx2 = runKMeans(X2, initial_centroids, int(i), num_iters)
    m, n = X2.shape
    X2_recovered = X2.copy()
    for j in range(1, int(i)+1):
        X2_recovered[(idx2==j).ravel(),:] = centroids2[j-1]
    X2_recovered = X2_recovered.reshape(rgb_img.shape[0], rgb_img.shape[1],3)
    plt.imshow(X2_recovered)
    plt.xticks([])
    plt.yticks([])
    plt.title('Image compression(K={})'.format(i))
    plt.savefig('./problem2/RGBimg_K{}.jpg'.format(i))
```

```python
from skimage import io
# for gray
image = io.imread('11HW2_PatchPanelsGrey.jpeg')

#Dimension of the original image
rows = image.shape[0]
cols = image.shape[1]

#Flatten the image
image = image.reshape(rows*cols, 3)

for i in k:
    #Implement k-means clustering to form k clusters
    kmeans = KMeans(n_clusters=int(i))
    kmeans.fit(image)

    #Replace each pixel value with its nearby centroid
    compressed_image = kmeans.cluster_centers_[kmeans.labels_]
    compressed_image = np.clip(compressed_image.astype('uint8'), 0, 255)

    #Reshape the image to original dimension
    compressed_image = compressed_image.reshape(rows, cols, 3)

    #Save and display output image
    io.imsave('./problem2/Grayimage_K{}.jpeg'.format(i), compressed_image)
    io.imshow(compressed_image)
    io.show()
```
